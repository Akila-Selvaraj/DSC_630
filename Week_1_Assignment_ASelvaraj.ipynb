```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
health_data_df=pd.read_csv('healthcare-dataset-stroke-data.csv')
health_data_df.head()
```

```python
stroke_data_df = health_data_df[health_data_df.stroke == 1]
other_data_df = health_data_df[health_data_df.stroke != 1]
print(stroke_data_df.info)
```

Summary of data:
This healthcare dataset has information of patients who are affected and not affected by stroke.This dataset has details like 
age, hypertension, heart disease, heart disease, and bmi.

From this dataset, I want to explore on these questions.

1. Does increase in avg glucose level cause stroke?
2. Does stroke affects only aged people?


```python
# Create a histogram or bar graph from your data.
# Histogram of age
n, bins, patches = plt.hist(x=health_data_df.age, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Histogram of age')
```

```python
#Histogram of avg_glucose_level
n, bins, patches = plt.hist(x=stroke_data_df.avg_glucose_level, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('mg/dL')
plt.ylabel('Frequency')
plt.title('Histogram of avg glucose level')
# plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```

```python
#Plotting countplot on gender to see which gender is most affected by stroke
import seaborn as sns
sns.countplot(x='gender',data=stroke_data_df)
```

```python
# Create a boxplot from your data.
fig, ax = plt.subplots()
ax.boxplot(stroke_data_df['avg_glucose_level'])
plt.title('Box plot of avg glucose level')
plt.show()
```

This box plot shows the avg_glucose_level of stroke patients. Box plot indicates that the median of avg glucose level of stroke 
patients is around 110. Lower quartile is around 90 and upper quartile is closer to 200.

```python
fig, ax = plt.subplots()
ax.boxplot(stroke_data_df['age'])
plt.show()
```

This box plot shows the age of stroke patients. Box plot indicates that the median of age of stroke 
patients is greater than 60. Lower quartile is around 60 and upper quartile is 80, and the median is 70.

```python
# Create a bivariate plot from your data.
```

```python
#create scatterplot of Age vs avg_glucose_level
plt.scatter(stroke_data_df.age, stroke_data_df.avg_glucose_level)
plt.title('Age vs avg_glucose_level')
plt.xlabel('Age')
plt.ylabel('Avg_glucose_level')
```

```python
sns.relplot(x='age', y='avg_glucose_level', hue='gender',  data=stroke_data_df)  
```

```python
sns.displot(stroke_data_df,x='age',y='bmi',hue='gender') 
```

Conclusion :
The plot shows that count of stroke patients increases with increase in the age. The plots plotted so far indicates that age is one of the crucial factors which influences stroke. Also, from the plots, we can say stroke mostly affects female gender compared to male. From the different plots plotted against avg glucose level, dependency of stroke with avg glucose level looks shady, and stroke has affected the patients with less glucose level as well. The influence of avg glucose level against stroke is less in the chosen dataaset.
